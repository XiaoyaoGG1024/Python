# 处理2016年

```python
data2016=pd.read_csv("./data/2016.csv",dtype=object)
```

# 数据清洗

## 修改数据类型

### 将登录时间设置为日期型数据

```python
data2016["LogonDate"]=pd.to_datetime(data2016["LogonDate"])
```

```python
data2016.dtypes
```

### 调整列最大宽度查看啊完整信息

```python
pd.set_option("max_colwidth",300)
pd.set_option('display.max_columns', 1000)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 1000)
```

```python
data2016
```

### 去除部分列

```python
dataNew=data2016.drop(columns=["Unnamed: 0","Unnamed: 0.1","id","IpCountry","IpProvince","IpCity","IpDistrict","IpIsp","IpType"])
```

```python
dataNew
```

```python
dataNew["netType"]=dataNew["netType"].astype(str)
```

```python
dataNew.dtypes
```

## 补全数据

### 补全网络类型

```python
for i in dataNew[dataNew["netType"]=="Unknown"].index:
    if "NetType/NON_NETWORK" in dataNew.loc[i,"UserAgent"]:
        dataNew.loc[i,"netType"]="NON_NETWORK"
    elif "NetType/MOBILE" in dataNew.loc[i,"UserAgent"]:
            dataNew.loc[i,"netType"]="MOBILE"
    elif "NetType/uninet" in dataNew.loc[i,"UserAgent"]:
            dataNew.loc[i,"netType"]="uninet"
    elif "NetType/4G" in dataNew.loc[i,"UserAgent"]:
        dataNew.loc[i,"netType"]="4G"
    elif "NetType/ctlte" in dataNew.loc[i,"UserAgent"]:
        dataNew.loc[i, "netType"] = "ctlte"
    elif "NetType/cmwap" in dataNew.loc[i,"UserAgent"]:
        dataNew.loc[i, "netType"] = "cmwap"
    elif "NetType/ctwap" in dataNew.loc[i, "UserAgent"]:
        dataNew.loc[i, "netType"] = "ctwap"
    elif "NetType/uniwap" in dataNew.loc[i,"UserAgent"]:
        dataNew.loc[i, "netType"] = "uniwap"
```

```python
dataNew[dataNew["netType"]=="Unknown"]
```

# 构建数据模型与数据可视化

## 用户网络类型饼图呈现
## pie(x,explode,lables,autopct,pctdistance,shadow,labeldistance,startangle)
#explode:每一块突出程度，len(x)
#labels:标签
#autopc:圆内文本
#pctdistance:默认0.6 大于1文本在园外
#labeldistance标签文本大于1在园外

## 按照netType分组记数

```python
group=dataNew.groupby("netType").count().sort_values(by="Browser").reset_index()
```

```python
group
```

## 画图

```python
plt.figure(figsize=(10,8),dpi=160)
plt.title("用户网络类型分布")
plt.pie(group["userid"],explode=[0.05]*len(group["userid"]),autopct="%1.1f%%",pctdistance=0.9)
plt.legend(labels=group["netType"],frameon=False)
plt.show()
```

## 保存2016年网络类型和总数

```python
group[["netType","userid"]]
```

```python
group[["netType","userid"]].to_csv("./data/2016count.csv")
```
