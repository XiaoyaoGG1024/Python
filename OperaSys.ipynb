# 获取数据

```python
import pandas as pd
```

```python
os2016=pd.read_csv("./data1/2016operaSys.csv")
```

```python
os2016
```

```python
os2017=pd.read_csv("./data1/2017operaSys.csv")
os2018=pd.read_csv("./data1/2018operaSys.csv")
os2019=pd.read_csv("./data1/2019operaSys.csv")
os2020=pd.read_csv("./data1/2020operaSys.csv")
os2021=pd.read_csv("./data1/2021operaSys.csv")
os2022=pd.read_csv("./data1/2022operaSys.csv")
```

```python
os2016.rename(columns={"userid":"2016"},inplace=True)
os2017.rename(columns={"userid":"2017"},inplace=True)
os2018.rename(columns={"userid":"2018"},inplace=True)
os2019.rename(columns={"userid":"2019"},inplace=True)
os2020.rename(columns={"userid":"2020"},inplace=True)
os2021.rename(columns={"userid":"2021"},inplace=True)
os2022.rename(columns={"userid":"2022"},inplace=True)
```

```python
os2022
```

# 合并数据

```python
os=pd.merge(os2016,os2017,how="outer",on="operaSys")
os
```

```python
osSys=pd.merge(os,os2018,how="outer",on="operaSys")
osSys=pd.merge(osSys,os2019,how="outer",on="operaSys")
osSys=pd.merge(osSys,os2020,how="outer",on="operaSys")
osSys=pd.merge(osSys,os2021,how="outer",on="operaSys")
osSys=pd.merge(osSys,os2022,how="outer",on="operaSys")
osSys
```

```python
#补全空值,将NaN填充为0
nt=osSys.fillna(0)
```

```python
osSys
```

```python
osSys.index=osSys["operaSys"]
osSys
```

```python
osSys.drop(columns="operaSys",inplace=True)
osSys
```
