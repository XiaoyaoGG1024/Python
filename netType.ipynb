## 获取数据

```python
#引入pandas库
import pandas as pd
```

```python
#读取数据，本案例读取的是csv格式文件
nt2016=pd.read_csv(r"data1/2016netType.csv")
```

```python
nt2016
```

```python
#修改列名
nt2016.rename(columns={"userid":"2016"},inplace=True)
```

```python
nt2016
```

```python
#读取2017年网络来源数据
nt2017=pd.read_csv(r"data1/2017netType.csv")
#修改列名
nt2017.rename(columns={"userid":"2017"},inplace=True)
```

```python
#读取2018年网络来源数据
nt2018=pd.read_csv(r"data1/2018netType.csv")
#修改列名
nt2018.rename(columns={"userid":"2018"},inplace=True)
```

```python
#读取2019年网络来源数据
nt2019=pd.read_csv(r"data1/2019netType.csv")
#修改列名
nt2019.rename(columns={"userid":"2019"},inplace=True)
```

```python
#读取2020年网络来源数据
nt2020=pd.read_csv(r"data1/2020netType.csv")
#修改列名
nt2020.rename(columns={"userid":"2020"},inplace=True)
```

```python
#读取2021年网络来源数据
nt2021=pd.read_csv(r"data1/2021netType.csv")
#修改列名
nt2021.rename(columns={"userid":"2021"},inplace=True)
```

```python
#读取2022年网络来源数据
nt2022=pd.read_csv(r"data1/2022netType.csv")
#修改列名
nt2022.rename(columns={"userid":"2022"},inplace=True)
```

```python
nt2022
```

```python
nt2016
```

```python
nt2017
```

## 数据清洗

```python
#合并数据，合并16、17的数据，使用pandas.merge()
netType=pd.merge(nt2016,nt2017,how="outer",on="netType")
```

```python
netType
```

```python
#合并数据
netType=pd.merge(netType,nt2018,how="outer",on="netType")
netType=pd.merge(netType,nt2019,how="outer",on="netType")
netType=pd.merge(netType,nt2020,how="outer",on="netType")
netType=pd.merge(netType,nt2021,how="outer",on="netType")
netType=pd.merge(netType,nt2022,how="outer",on="netType")
```

```python
#查看合并后的数据
netType
```

```python
#补全空值,将NaN填充为0
nt=netType.fillna(0)
```

```python
#查看补全空值后的数据
nt
```

行列转换

```python
#将netType变成index索引
nt.index=nt["netType"]
```

```python
#查看数据
nt
```

```python
#删除columns中的netType
nt.drop(columns="netType",inplace=True)
```

```python
#查看nt
nt
```

```python
#查看index
nt.index
```

```python
#查看columns
nt.columns
```

```python
#交换行和列
nt=pd.DataFrame(nt.values.T,columns=nt.index,index=nt.columns)
```

```python
#查看nt
nt
```

## 数据可视化

```python
#绘制4G、WIFI网络来源趋势图
#引入matplotlib.pyplot库
import matplotlib.pyplot as plt
```

```python
#定义画布
plt.figure(figsize=(8,7),dpi=180)
#设置字体
plt.rcParams["font.sans-serif"]="SimHei"
#设置标题
plt.title("4G用户")
#画图
plt.plot(nt.index,nt["4G"])

plt.show()
```

# 绘制子图

```python
fig=plt.figure(figsize=(10,10),dpi=180)
axes=fig.subplots(3,3)
#2016
axes[0,0].set_title("2016",fontsize=9)
axes[0,0].pie(nt2016["2016"],labels=nt2016["netType"],textprops={"fontsize":7},autopct="%1.1f%%")
#2017
axes[0,1].set_title("2017",fontsize=9)
axes[0,1].pie(nt2017["2017"],labels=nt2017["netType"],textprops={"fontsize":7},autopct="%1.1f%%")
#2018
axes[0,2].set_title("2018",fontsize=9)
axes[0,2].pie(nt2018["2018"],labels=nt2018["netType"],textprops={"fontsize":7},autopct="%1.1f%%")
#2019
axes[1,0].set_title("2019",fontsize=9)
axes[1,0].pie(nt2019["2019"],labels=nt2019["netType"],textprops={"fontsize":7},autopct="%1.1f%%")
#2020
axes[1,1].set_title("2020",fontsize=9)
axes[1,1].pie(nt2020["2020"],labels=nt2020["netType"],textprops={"fontsize":7},autopct="%1.1f%%")
#2021
axes[1,2].set_title("2021",fontsize=9)
axes[1,2].pie(nt2021["2021"],labels=nt2021["netType"],textprops={"fontsize":7},autopct="%1.1f%%")
#2022
axes[2,0].set_title("2022",fontsize=9)
axes[2,0].pie(nt2022["2022"],labels=nt2022["netType"],textprops={"fontsize":7},autopct="%1.1f%%")
#4G
axes[2,1].set_title("4G",fontsize=9)
axes[2,1].plot(nt.index,nt["4G"])
#WIFI
axes[2,2].set_title("WIFI",fontsize=9)
axes[2,2].plot(nt.index,nt["WIFI"])
plt.show()
```
